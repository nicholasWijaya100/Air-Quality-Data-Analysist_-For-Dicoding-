```python
# function to calculate AQI for PM2.5
def calculate_aqi_pm25(pm25):
    # AQI breakpoints and corresponding sub-indices
    breakpoints = [0, 12, 35.4, 55.4, 150.4, 250.4, 350.4, 500.4]
    sub_indices = [0, 50, 100, 150, 200, 300, 400, 500]
    aqi_range = []

    # Find the appropriate AQI range for the given PM2.5 concentration
    for i in range(len(breakpoints) - 1):
        if breakpoints[i] <= pm25 <= breakpoints[i + 1]:
            aqi_range = sub_indices[i:i + 2]
            break
    else:
        # Handle the case where no suitable AQI range is found
        aqi_range = [0, 0]

    # Calculate the sub-index for PM2.5
    aqi_value = ((aqi_range[1] - aqi_range[0]) / (breakpoints[i + 1] - breakpoints[i])) * (pm25 - breakpoints[i]) + aqi_range[0]

    return round(aqi_value)

combined_df['AQI_PM25'] = combined_df['PM2.5'].apply(calculate_aqi_pm25)
print(combined_df.head())
```
